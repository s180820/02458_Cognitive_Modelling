```python
# calculate the perceived stimulus intensity, Ip, for physical intensities 
# Is = 1, 2, . . . , 10 using Stevenâ€™s law.

import numpy as np
import matplotlib.pyplot as plt

# define the function
def Stevens(Is,a):
    return 10*Is**(a)

# define the range of Is
Is = np.arange(1,11)

# calculate the perceived stimulus intensity
Ip = Stevens(Is,0.33)

# plot the results
plt.plot(Is, Ip, 'o-')
plt.xlabel('Physical Intensity')
plt.ylabel('Perceived Intensity')
plt.title('Steven\'s Law')
plt.show()
```

```python
# This simulates an observer that rates the perceived intensity according 
# to Stevensâ€™ law. Fit Fechnerâ€™s law to the simulated data. 
# Note that Fechnerâ€™s law is linear with respect to Is.

# define the function
def fechner(Is,c, I0):
    return (1/c)*np.log(Is/I0)

from scipy.optimize import curve_fit

param, cov = curve_fit(fechner, Is, Ip)
fit_fech = fechner(Is, param[0], param[1])

print("Optimal Weber fraction", param[0])
print("I0:", param[1])

# plot the results
plt.plot(Is, Ip, 'o-',label='Steven\'s Law')
plt.plot(Is, fit_fech, 'o-', label='Fitted Fechner\'s Law (with a=0.33)')
plt.xlabel('Physical Intensity')
plt.ylabel('Perceived Intensity')
#plt.title('Fitted Fechner\'s Law (with a=0.33)')
plt.legend()
plt.show()
```

```python
Ip = Stevens(Is, 3.3)

param, cov = curve_fit(fechner, Is, Ip, maxfev=5000)
fit_fech = fechner(Is, param[0], param[1])

print("Optimal Weber fraction:", param[0])
print("I0:", param[1])
# plot the results
plt.plot(Is, Ip, 'o-',label='Steven\'s Law')
plt.plot(Is, fit_fech, 'o-', label='Fitted Fechner\'s Law (with a=3.3)')
plt.xlabel('Physical Intensity')
plt.ylabel('Perceived Intensity')
#plt.title('Fitted Fechner\'s Law (with a=3.3)')
plt.legend()
plt.show()
```
